```python
#problem1
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Step 2: Load dataset
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/WHR20245CS037_a1bfcb8e-f3d5-49cf-9072-5f0f6923bbef_264243_.csv")
# Display first 10 rows
df.head(10)
```

```python
# Number of rows and columns
print("Shape of dataset:", df.shape)

# Columns and their data types
print(df.dtypes)
```

```python
# Mean, Median, Standard Deviation of Score
mean_score = df['score'].mean()
median_score = df['score'].median()
std_score = df['score'].std()

print("Mean Score:", mean_score)
print("Median Score:", median_score)
print("Standard Deviation:", std_score)

# Country with highest and lowest happiness score
highest_country = df.loc[df['score'].idxmax(), 'Country name']
lowest_country = df.loc[df['score'].idxmin(), 'Country name']

print("Highest Happiness Score Country:", highest_country)
print("Lowest Happiness Score Country:", lowest_country)

```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Filter countries with score > 7.5
high_score_df = df[df['score'] > 7.5]

# Sort by Log GDP per capita in descending order
top_gdp_df = high_score_df.sort_values(by='Log GDP per capita', ascending=False).head(10)

# Display the top 10 rows
top_gdp_df
```

```python
# Categorize countries by happiness score
def happiness_category(score):
    if score < 4:
        return "Low"
    elif 4 <= score <= 6:
        return "Medium"
    else:
        return "High"

# Apply the function to the 'score' column
df['Happiness Category'] = df['score'].apply(happiness_category)

# Display the first 10 rows to verify
df.head(10)
```

```python
## Bar plot: Top 10 happiest countries
top10_happy = df.sort_values(by='score', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='score', y='Country name', data=top10_happy, palette='viridis')
plt.title("Top 10 Happiest Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Country")
plt.show()
```

```python
# Line plot: Top 10 unhappiest countries
bottom10_happy = df.sort_values(by='score').head(10)
plt.figure(figsize=(10,6))
sns.lineplot(x='Country name', y='score', data=bottom10_happy, marker='o')
plt.title("Top 10 Unhappiest Countries")
plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("Happiness Score")
plt.show()
```

```python
# Histogram of Happiness Scores
plt.figure(figsize=(8,6))
sns.histplot(df['score'], bins=15, kde=True, color='skyblue')
plt.title("Distribution of Happiness Scores")
plt.xlabel("Happiness Score")
plt.ylabel("Count")
plt.show()

```

```python
#problem2
# Step 1: Define the list of South Asian countries
south_asian_countries = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

# Step 2: Filter only those countries
south_asia_df = df[df["Country name"].isin(south_asian_countries)]

# Step 3: Save filtered dataset (optional)
south_asia_df.to_csv("South_Asia_Happiness.csv", index=False)

# Display the South Asia DataFrame
south_asia_df
```

```python
# Step 1: Create new column "Composite Score"
south_asia_df["Composite Score"] = (
    0.40 * south_asia_df["Log GDP per capita"]
    + 0.30 * south_asia_df["Social support"]
    + 0.30 * south_asia_df["Healthy life expectancy"]
)

# Step 2: Rank countries based on Composite Score
south_asia_df = south_asia_df.sort_values(by="Composite Score", ascending=False)

# Step 3: Visualize top 5 countries
top5_composite = south_asia_df.head(5)
plt.figure(figsize=(8,5))
sns.barplot(x="Composite Score", y="Country name", data=top5_composite, palette="plasma")
plt.title("Top 5 South Asian Countries by Composite Score")
plt.xlabel("Composite Score")
plt.ylabel("Country")
plt.show()

# Step 4: Compare Composite Score vs Happiness Score
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="Composite Score", y="score",
    data=south_asia_df, s=100, color="orange"
)
plt.title("Composite Score vs Happiness Score  South Asia")
plt.xlabel("Composite Score")
plt.ylabel("Happiness Score")
plt.show()
```

```python
# Step 1: Compute IQR for score
Q1 = south_asia_df["score"].quantile(0.25)
Q3 = south_asia_df["score"].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = south_asia_df[(south_asia_df["score"] < lower_bound) | (south_asia_df["score"] > upper_bound)]

print("Outlier Countries based on Happiness Score:")
print(outliers[["Country name", "score", "Log GDP per capita"]])

# Step 3: Scatter plot highlighting outliers
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="Log GDP per capita", y="score",
    data=south_asia_df, color="blue", s=100, label="Normal"
)
sns.scatterplot(
    x="Log GDP per capita", y="score",
    data=outliers, color="red", s=120, label="Outlier"
)
plt.title("Outlier Detection  South Asian Countries")
plt.xlabel("Log GDP per Capita")
plt.ylabel("Happiness Score")
plt.legend()
plt.show()
```

```python
# Step 1: Calculate correlations
corr_freedom = south_asia_df["Freedom to make life choices"].corr(south_asia_df["score"])
corr_generosity = south_asia_df["Generosity"].corr(south_asia_df["score"])

print("Correlation (Freedom vs Score):", corr_freedom)
print("Correlation (Generosity vs Score):", corr_generosity)

# Step 2: Scatter plots with trendlines
plt.figure(figsize=(8,5))
sns.regplot(
    x="Freedom to make life choices", y="score",
    data=south_asia_df, scatter_kws={"color":"green"}, line_kws={"color":"black"}
)
plt.title("Freedom to Make Life Choices vs Happiness Score (South Asia)")
plt.xlabel("Freedom to Make Life Choices")
plt.ylabel("Happiness Score")
plt.show()

plt.figure(figsize=(8,5))
sns.regplot(
    x="Generosity", y="score",
    data=south_asia_df, scatter_kws={"color":"purple"}, line_kws={"color":"black"}
)
plt.title("Generosity vs Happiness Score (South Asia)")
plt.xlabel("Generosity")
plt.ylabel("Happiness Score")
plt.show()
```

```python
# Step 1: Add new column
south_asia_df["GDP-Score Gap"] = south_asia_df["Log GDP per capita"] - south_asia_df["score"]

# Step 2: Rank countries by gap
gap_asc = south_asia_df.sort_values(by="GDP-Score Gap").head(3)
gap_desc = south_asia_df.sort_values(by="GDP-Score Gap", ascending=False).head(3)

print("Top 3 Countries with Largest Positive Gap:")
print(gap_desc[["Country name", "GDP-Score Gap"]])
print("\nTop 3 Countries with Largest Negative Gap:")
print(gap_asc[["Country name", "GDP-Score Gap"]])

# Step 3: Visualize
plt.figure(figsize=(8,5))
sns.barplot(x="GDP-Score Gap", y="Country name", data=gap_desc, color="green")
plt.title("Top 3 Countries with Positive GDPScore Gap (South Asia)")
plt.xlabel("GDP  Score Gap")
plt.ylabel("Country")
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(x="GDP-Score Gap", y="Country name", data=gap_asc, color="red")
plt.title("Top 3 Countries with Negative GDPScore Gap (South Asia)")
plt.xlabel("GDP  Score Gap")
plt.ylabel("Country")
plt.show()
```

```python
# Define the list of Middle Eastern countries
middle_east_countries = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
    "Kuwait", "Lebanon", "Oman", "Palestine",
    "Qatar", "Saudi Arabia", "Syria",
    "United Arab Emirates", "Yemen"
]

# Filter dataset for Middle East countries
middle_east_df = df[df["Country name"].isin(middle_east_countries)]

# Optional: Save to CSV
middle_east_df.to_csv("Middle_East_Happiness.csv", index=False)

# Display dataset
middle_east_df.head()
```

```python
# Calculate mean and standard deviation of happiness score for both regions
mean_sa = south_asia_df["score"].mean()
std_sa  = south_asia_df["score"].std()
mean_me = middle_east_df["score"].mean()
std_me  = middle_east_df["score"].std()

print("South Asia  Mean:", round(mean_sa, 2), "| Std Dev:", round(std_sa, 2))
print("Middle East  Mean:", round(mean_me, 2), "| Std Dev:", round(std_me, 2))

# Identify which region is happier
if mean_sa > mean_me:
    print("\nSouth Asia has a higher average happiness score.")
else:
    print("\nMiddle East has a higher average happiness score.")
```

```python
# Top and bottom 3 for both regions
top3_sa    = south_asia_df.nlargest(3, "score")
bottom3_sa = south_asia_df.nsmallest(3, "score")

top3_me    = middle_east_df.nlargest(3, "score")
bottom3_me = middle_east_df.nsmallest(3, "score")

# Plot comparison
plt.figure(figsize=(9,5))
sns.barplot(x="score", y="Country name", data=top3_sa, color="green", label="South Asia Top 3")
sns.barplot(x="score", y="Country name", data=top3_me, color="orange", label="Middle East Top 3")
plt.title("Top 3 Countries  South Asia vs Middle East")
plt.xlabel("Happiness Score")
plt.legend()
plt.show()
```

```python
metrics = ["Log GDP per capita", "Social support", "Healthy life expectancy"]

avg_sa = south_asia_df[metrics].mean()
avg_me = middle_east_df[metrics].mean()

compare_df = pd.DataFrame({"Metric": metrics, "South Asia": avg_sa.values, "Middle East": avg_me.values})

# Grouped bar chart
compare_df.plot(
    x="Metric", kind="bar", figsize=(8,5),
    color=["skyblue", "salmon"], rot=15
)
plt.title("Average Metric Comparison  South Asia vs Middle East")
plt.ylabel("Average Value")
plt.show()
```

```python
# South Asia
range_sa = south_asia_df["score"].max() - south_asia_df["score"].min()
cv_sa    = std_sa / mean_sa

# Middle East
range_me = middle_east_df["score"].max() - middle_east_df["score"].min()
cv_me    = std_me / mean_me

print("South Asia  Range:", round(range_sa,2), "| CV:", round(cv_sa,2))
print("Middle East  Range:", round(range_me,2), "| CV:", round(cv_me,2))

if cv_sa > cv_me:
    print("\nSouth Asia has greater variability in happiness.")
else:
    print("\nMiddle East has greater variability in happiness.")
```

```python
corr_sa_freedom = south_asia_df["Freedom to make life choices"].corr(south_asia_df["score"])
corr_sa_generosity = south_asia_df["Generosity"].corr(south_asia_df["score"])

corr_me_freedom = middle_east_df["Freedom to make life choices"].corr(middle_east_df["score"])
corr_me_generosity = middle_east_df["Generosity"].corr(middle_east_df["score"])

print("South Asia  Freedom Corr:", round(corr_sa_freedom,2), "| Generosity Corr:", round(corr_sa_generosity,2))
print("Middle East  Freedom Corr:", round(corr_me_freedom,2), "| Generosity Corr:", round(corr_me_generosity,2))

# Scatter plots
plt.figure(figsize=(8,5))
sns.scatterplot(x="Freedom to make life choices", y="score", data=south_asia_df, label="South Asia", color="green")
sns.scatterplot(x="Freedom to make life choices", y="score", data=middle_east_df, label="Middle East", color="orange")
plt.title("Freedom vs Happiness Score")
plt.legend()
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x="Generosity", y="score", data=south_asia_df, label="South Asia", color="purple")
sns.scatterplot(x="Generosity", y="score", data=middle_east_df, label="Middle East", color="red")
plt.title("Generosity vs Happiness Score")
plt.legend()
plt.show()
```

```python
# Combine both regions for boxplot comparison
combined_df = pd.concat([
    south_asia_df.assign(Region="South Asia"),
    middle_east_df.assign(Region="Middle East")
])

plt.figure(figsize=(8,5))
sns.boxplot(x="Region", y="score", data=combined_df, palette="Set2")
plt.title("Happiness Score Distribution  South Asia vs Middle East")
plt.xlabel("Region")
plt.ylabel("Happiness Score")
plt.show()
```
